```python
import numpy as np
import numpy.linalg as lin
print(np.version.full_version)
from scipy import linalg
import math
import matplotlib.pyplot as plt
```

```python
t=np.array(list(range(0,13,2)))
f=np.array([3,4,5,5.5,6.5,7,8])
plt.plot(t,f,'x')
A=np.array([[7,42],[42,364]])
b=np.array([39,279])
beta=lin.solve(A,b)
print(beta,45/112)
plt.plot([0,12],np.polyval(beta[::-1],[0,12]))
print(np.polyval(beta[::-1],[20]))
(50-beta[0])/beta[1]
```

```python

A=np.array([[1,v,math.exp(v)] for v in t])

B=np.dot(A.transpose(),A)
b=np.dot(A.transpose(),f)
beta=lin.solve(B,b)
plt.plot(t,f,'x')
x=np.linspace(0,12,100)
plt.plot(x,np.polyval(beta[::-1],x))
```

```python
t=[-2,-1,0,2]
f=[-31,-7,-1,5]
beta=np.polyfit(t,f,3)
print(beta)
plt.plot(t,f,'x')
x=np.linspace(-2,2)
plt.plot(x,np.polyval(beta,x))
```

```python
p=[2,-9,24,-31]

def horner(x,p,t):
    px=0
    for i in range(len(p)):
        px=px*(x-t[i])+p[i]
    return px

plt.plot(t,f,'x')
plt.plot(x,[horner(v,p,t[::-1]) for v in x])
```
